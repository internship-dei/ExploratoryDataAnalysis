```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Salesstore.csv',sep = '\t')
df
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plot
import seaborn as sb
```

```python
plot.scatter(df.index, df['Ship_Mode'])
plot.show()
# People tend to prefer Regular air as ship mode
```

```python
plot.scatter(df.index, df['Region'])
plot.show()

# Prarie has least amount of orders while West has most 
```

```python
plot.scatter(df.index, df['Customer_Segment'])
plot.show()
# The plot shows the quantity of orders made by each customer segment.
```

```python
plot.scatter(df.index, df['Product_Container'])
plot.show()
# Most of the packages came in Small box
```

```python
sb.scatterplot(x=df['Profit'], y=df['Sales'], hue=df['Region'])
# Data showed more profit margin in Northwest territories while less in Ontario 
```

```python
sb.scatterplot(x=df['Region'], y=df['Profit'], hue=df['Customer_Segment'])
# Consumers are giving higher profits profit 
```

```python
# sb.scatterplot(x = df['Region'],y = df['Profit'],hue=df['Order_Quantity'])
sb.violinplot(x=df['Region'], y=df['Profit'])
```

```python
ax = sb.violinplot(x=df['Region'], y=df['Profit'],
                   hue=df['Ship_Mode'], palette="muted", split=True)

# Nunavut, ontario people preferred Express air as shipping mode, West and Atlantic preferred Regular Air 
```

```python
sb.histplot(data=df, x="Customer_Segment")
# more no of corporate orders and so on
```

```python
sb.histplot(data=df, x="Customer_Segment", y="Region")
# Darker regions show more no of orders from a customer segment 
```

```python
ax = sb.boxplot(x="Region", y="Order_Quantity", data=df)

# Median of the quantity of orders was plotted, ontario data shows more order quantities 
```

```python
sb.lineplot(data=df,x="Sales", y="Region", hue="Customer_Segment",markers=True)
```
