```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("Salesstore.csv", error_bad_lines=False,sep="\t")
df.head()
```

```python
df.dtypes
```

```python
df.describe()
```

# **Scatter Plot**

```python
sns.scatterplot(data=df, x="Order_Quantity", y="Profit")
```

```python
sns.scatterplot(df.index, df['Ship_Mode'])
```

```python
sns.scatterplot(df.index, df['Region'])
```

```python
sns.scatterplot(df.index, df['Customer_Segment'])
```

```python
sns.scatterplot(x=df['Sales'], y=df['Profit'], hue=df['Region'])
```

# **Box Plot**

```python
ax = sns.boxplot(x=df["Order_Quantity"])
```

```python
ax = sns.boxplot(x="Region", y="Order_Quantity", data=df)
```

# **Line Plot**

```python
sns.lineplot(data=df, x="Order_Quantity", y="Profit")
```

```python
sns.lineplot(data=df, x="Sales", y="Region", hue="Customer_Segment")
```

# **Violin Plot**

```python
ax = sns.violinplot(x=df["Profit"])
```

```python
ax = sns.violinplot(x=df['Region'], y=df['Profit'], hue=df['Ship_Mode'])
```

# **Bar Plot**

```python
sns.histplot(data=df, x="Profit")
```

```python
sns.histplot(data=df, y="Profit")
```

```python
sns.histplot(data=df, x="Customer_Segment")
```

## **Pair Plot**

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="Profit")
```
